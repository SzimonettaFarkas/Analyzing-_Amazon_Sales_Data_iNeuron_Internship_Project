```python
# creating a copy of a dataset 
data1 = data.copy()
data1.head()
```

```python
# dropping useless columns
data1.drop(data1.columns[[20,21]], axis = 1, inplace = True)
```

```python
data1.columns
```

```python
# sum of missing values in columns
data1.isna().sum()
```

```python
# unique values in Item Class column
data1['Item Class']. unique()
```

```python
# number of particular values
data1.groupby('Item Class').size()
```

```python
# filling missing values
data1['Item Class'].fillna("P01", inplace = True)
```

```python
# fixing wrong values
data1['Item Class'] = data1['Item Class'].replace('PO1', 'P01')
```

```python
data1.groupby('Item Class').size()
```

```python
# removing null values from the dataset
data1.dropna(subset=['Discount Amount', 'Item Number', 'Sales Price'], inplace = True)
```

```python
data1.isna().sum()
```

```python
data1.describe().round() # statistical description
```

```python
# creating new columns: year, month, quarter, day
data1['Year'] = data1['Invoice Date'].dt.year
data1['Month'] = data1['Invoice Date'].dt.month
data1['Quarter'] = data1['Invoice Date'].dt.quarter
data1['Day'] = data1['Invoice Date'].dt.day
```

```python
data1.head()
```

```python
# creating a dataframe with the numerical columns
dataset = data1[['CustKey', 'Discount Amount','Invoice Number', 'Line Number', 'List Price', 'Order Number', 'Sales Amount Based on List Price',
                 'Sales Amount', 'Sales Cost Amount', 'Sales Margin Amount', 'Sales Price', 'Sales Quantity', 'Sales Rep']]
```

```python
dataset.head()
```

```python
# checking the correlation with the help of a heatmap
plt.figure(figsize=(15,10))
sns.heatmap(dataset.corr(method='pearson'), annot = True, vmin = -1, vmax = 1, cmap = 'YlOrRd')
```

```python
# INSIGHTS

# STRONG POSITIVE RELATIONSHIP BETWEEN: 
    # List Price - Sales Price
    # Sales Amount Based on List Price - Sales Amount
    # Sales Amount Based on List Price - Sales Cost Amount
    # Sales Amount Based on List Price - Discount Amount
    # Sales Amount Based on List Price - Sales Margin Amount
    # Sales Amount Based on List Price - Sales Quantity
    # Sales Amount - Sales Cost Amount
    # Sales Amount - Sales Margin Amount
    # Sales Amount - Sales Quantity
    # Sales Cost Amount - Sales Margin Amount 
    # Sales Cost Amount - Sales Quantity
    # Sales Margin Amount - Sales Quantity
```
